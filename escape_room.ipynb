```python
# حالة اللعبة - Game State
game_state = {
    "current_room": "game_room",
    "keys_collected": [],
    "unlocked_doors": []
}

# تعريف الغرف
rooms = {
    "game_room": {
        "description": "You are in a game room",
        "objects": ["couch", "piano", "door_a"],
        "items_in_objects": {"piano": "key_a"}
    },
    "bedroom_1": {
        "description": "You are in Bedroom 1.",
        "objects": ["queen_bed", "door_a", "door_b", "door_c"],
        "items_in_objects": {"queen_bed": "key_b"}
    },
    "bedroom_2": {
        "description": "You are in Bedroom 2.",
        "objects": ["double_bed", "dresser", "door_b"],
        "items_in_objects": {"double_bed": "key_c", "dresser": "key_d"}
    },
    "living_room": {
        "description": "You are in the living room.",
        "objects": ["dining_table", "door_c", "door_d"],
        "items_in_objects": {}
    }
}

doors = {
    "door_a": {"rooms": ["game_room", "bedroom_1"], "key": "key_a"},
    "door_b": {"rooms": ["bedroom_1", "bedroom_2"], "key": "key_b"},
    "door_c": {"rooms": ["bedroom_1", "living_room"], "key": "key_c"},
    "door_d": {"rooms": ["living_room", "outside"], "key": "key_d"}
}

def start_game():
    print("--- Welcome to the Escape Room Game! ---")
    play_game()

def play_game():
    while True:
        current_room_id = game_state["current_room"]
        current_room = rooms[current_room_id]
        
        print("\n" + current_room["description"])
        command = input("What would you like to do? (explore / examine [object] / go [door]): ").lower().strip()

        if command == "explore":
            explore_room(current_room)
            
        elif command.startswith("examine "):
            parts = command.split(" ")
            if len(parts) > 1:
                item_name = parts[1]
                examine_item(game_state, item_name, current_room["objects"], current_room["items_in_objects"])
            else:
                print("Examine what? please specify an object.")

        elif command.startswith("go "):
            commandpartList = command.split(" ")
            if len(commandpartList) > 1:
                door_name = commandpartList[1]
                go_to_door(game_state, door_name, current_room["objects"])
            else:
                print("Go where? please specify a door.")

        elif command == "quit":
            print("Thanks for playing!")
            break
        else:
            print("Invalid command. Try: explore, examine [object], go [door], or quit.")

def explore_room(room):
    print(f"You look around and see: {', '.join(room['objects'])}")

def examine_item(game_state, item_name, objects_in_room, items_in_objects):
    if item_name not in objects_in_room:
        print(f"The {item_name} is not in this room.")
        return

    print(f"You examine the {item_name}...")
    item_found = items_in_objects.get(item_name)
    
    if item_found:
        print(f"Success! You found: {item_found}!")
        if item_found not in game_state["keys_collected"]:
            game_state["keys_collected"].append(item_found)
    else:
        print(f"There is nothing hidden in the {item_name}.")

def go_to_door(game_state, door_name, objects_in_room):
    if door_name not in objects_in_room:
        print(f"There is no {door_name} here.")
        return

    if door_name not in doors:
        print(f"Can not find {door_name} in the game.")
        return

    door_info = doors[door_name]
    required_key = door_info["key"]
    if door_name in game_state["unlocked_doors"] or required_key in game_state["keys_collected"]:
        if door_name not in game_state["unlocked_doors"]:
            print(f"You use the {required_key} to unlock {door_name}.")
            game_state["unlocked_doors"].append(door_name)

        next_room = [r for r in door_info["rooms"] if r != game_state["current_room"]][0]
        
        if next_room == "outside":
            print("Congratulations! You've escaped the room!")
            exit()
        else:
            game_state["current_room"] = next_room
            print(f"You pass through {door_name} and enter the {next_room}.")
    else:
        print(f"The {door_name} is locked. You need the {required_key} to open it.")

if __name__ == "__main__":
    start_game()
```
