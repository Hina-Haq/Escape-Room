```python
# CELL 2: RELATIONSHIPS AND GAME STATE 

object_relations = {
    "game room": [
        couch,
        piano,
        wardrobe_1,
        chair_1,
        chair_2,
        coffee_table_1,
        door_a,
        door_b
    ],
    "piano": [key_a],

    "door a": [game_room, room_2],
    "door b": [game_room, bathroom],

    "game lounge": [
        couch_2,
        pool_table,
        mini_fridge,
        bathroom_door,
        door_c
    ],
    "mini fridge": [key_c],

    "bathroom door": [room_2, bathroom],
    "door c": [room_2, bedroom],

    "bathroom": [
        sink,
        mirror,
        toilet,
        bathroom_door
    ],

    "bedroom": [
        queen_bed,
        wardrobe_2,
        dressing_table,
        door_d,
        door_e
    ],
    "bedroom wardrobe": [key_d],
    "dressing table": [key_e],

    "door d": [bedroom, game_room],
    "door e": [bedroom, living_room],

    "living room": [
        sofa,
        coffee_table_2,
        tv,
        work_desk,
        computer,
        door_f
    ],
    "computer": [key_f],

    "door f": [living_room, outside],

    "outside": []
}

INIT_GAME_STATE = {
    "current_room": game_room,
    "keys_collected": [],
    "target_room": outside
}

game_state = INIT_GAME_STATE.copy()
```

```python
# CELL 3: FUNCTIONS 

def linebreak():
    print("\n")


def start_game():
    game_state["current_room"] = game_room
    game_state["keys_collected"] = []

    print("You wake up on a couch in a strange house.")
    print("You do not know how you got here.")
    print("You must explore the rooms, find keys, unlock doors, and escape.")
    linebreak()

    play_game()


def play_game():
    while True:
        current_room = game_state["current_room"]

        if current_room == game_state["target_room"]:
            print("Congrats! You escaped the house!")
            break

        print("You are now in:", current_room["name"])
        print("Available commands: explore, examine, inventory, quit")
        command = input("What would you like to do? ").strip().lower()

        if command == "explore":
            explore_room(current_room)

        elif command == "examine":
            item_name = input("What would you like to examine? ").strip().lower()
            examine_item(item_name)

        elif command == "inventory":
            show_inventory()

        elif command == "quit":
            print("Game ended.")
            break

        else:
            print("Invalid command. Try: explore, examine, inventory, or quit.")

        linebreak()


def explore_room(room):
    items = [item["name"] for item in object_relations[room["name"]]]
    print("You explore the room and find:", ", ".join(items))


def show_inventory():
    if len(game_state["keys_collected"]) == 0:
        print("You do not have any keys yet.")
    else:
        key_names = [key["name"] for key in game_state["keys_collected"]]
        print("You have:", ", ".join(key_names))


def get_next_room_of_door(door, current_room):
    connected_rooms = object_relations[door["name"]]

    for room in connected_rooms:
        if room != current_room:
            return room


def examine_item(item_name):
    current_room = game_state["current_room"]
    found_item = None

    for item in object_relations[current_room["name"]]:
        if item["name"].lower() == item_name:
            found_item = item
            break

    if found_item is None:
        print("That item is not in this room.")
        return

    print("You examine", found_item["name"] + ".")

    if found_item["type"] == "door":
        open_door(found_item, current_room)

    else:
        inspect_furniture(found_item)


def open_door(door, current_room):
    have_key = False

    for key in game_state["keys_collected"]:
        if key["target"] == door:
            have_key = True
            break

    if have_key:
        print("You unlock", door["name"], "with the correct key.")
        next_room = get_next_room_of_door(door, current_room)
        game_state["current_room"] = next_room
        print("You move to", next_room["name"] + ".")
    else:
        print("This door is locked. You do not have the correct key.")


def inspect_furniture(item):
    item_name = item["name"]

    if item_name in object_relations and len(object_relations[item_name]) > 0:
        found_object = object_relations[item_name].pop()
        game_state["keys_collected"].append(found_object)
        print("You find", found_object["name"] + "!")
    else:
        print("There is nothing useful here.")
```
