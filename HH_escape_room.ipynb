```python
# ---------- RELATIONSHIPS ----------

object_relations = {
    # game room
    "game room": [
        couch,
        piano,
        wardrobe_1,
        chair_1,
        chair_2,
        coffee_table_1,
        door_a,
        door_b,
    ],
    "piano": [key_a],

    # door connections
    "door a": [game_room, room_2],
    "door b": [game_room, bathroom],   # wrong door at first, no matching key

    # room 2
    "game lounge": [
        couch_2,
        pool_table,
        mini_fridge,
        bathroom_door,
        door_c,
    ],
    "mini fridge": [key_c],
    "bathroom door": [room_2, bathroom],
    "door c": [room_2, bedroom],

    # bathroom
    "bathroom": [
        sink,
        mirror,
        toilet,
        bathroom_door,
    ],

    # bedroom
    "bedroom": [
        queen_bed,
        wardrobe_2,
        dressing_table,
        door_d,
        door_e,
    ],
    "bedroom wardrobe": [key_d],
    "dressing table": [key_e],
    "door d": [bedroom, game_room],
    "door e": [bedroom, living_room],

    # living room
    "living room": [
        sofa,
        coffee_table_2,
        tv,
        work_desk,
        computer,
        keypad_door,
    ],
    "keypad door": [living_room, outside],

    # outside
    "outside": [keypad_door],
} 
```

```python
# INITIAL GAME STATE 

INIT_GAME_STATE = {
    "current_room": game_room,
    "keys_collected": [],
    "target_room": outside,
}

game_state = INIT_GAME_STATE.copy()  

# INIT_GAME_STATE = original starting blueprint.

# game_state = the version the game changes while the player plays.
```

```python
# ---------- HELPER FUNCTION ----------

def linebreak():
    print("\n\n")


# ---------- GAME FUNCTIONS ----------

def start_game():
    print("You wake up on a couch and find yourself in a strange house with no windows.")
    print("You don't remember how you got here.")
    print("You feel danger approaching, and you must escape the house.")
    linebreak()
    play_room(game_state["current_room"])


def play_room(room):
    game_state["current_room"] = room

    if game_state["current_room"] == game_state["target_room"]:
        print("Congrats! You escaped the house!")
        return
    else:
        print("You are now in " + room["name"])
        intended_action = input("What would you like to do? Type 'explore' or 'examine': ").strip().lower()

        if intended_action == "explore":
            explore_room(room)
            linebreak()
            play_room(room)

        elif intended_action == "examine":
            item_name = input("What would you like to examine? ").strip().lower()
            examine_item(item_name)

        else:
            print("Not sure what you mean. Type 'explore' or 'examine'.")
            linebreak()
            play_room(room)


def explore_room(room):
    items = [i["name"] for i in object_relations[room["name"]]]
    print("You explore the room. This is " + room["name"] + ". You find " + ", ".join(items))


def get_next_room_of_door(door, current_room):
    connected_rooms = object_relations[door["name"]]
    for room in connected_rooms:
        if room != current_room:
            return room
```

```python
def examine_item(item_name):
    current_room = game_state["current_room"]
    next_room = None
    output = None

    for item in object_relations[current_room["name"]]:
        if item["name"].lower() == item_name:
            output = "You examine " + item["name"] + ". "

            # if the item is a normal door
            if item["type"] == "door":
                have_key = False

                for key in game_state["keys_collected"]:
                    if key["target"] == item:
                        have_key = True

                if have_key:
                    output += "You unlock it with a key you have."
                    next_room = get_next_room_of_door(item, current_room)
                else:
                    output += "It is locked but you don't have the correct key."

            # if the item is furniture or something that may hide a key
            else:
                if item["name"] in object_relations and len(object_relations[item["name"]]) > 0:
                    item_found = object_relations[item["name"]].pop()
                    game_state["keys_collected"].append(item_found)
                    output += "You find " + item_found["name"] + "."
                else:
                    output += "There isn't anything interesting about it."

            print(output)
            break

    if output is None:
        print("That item is not in this room.")

    if next_room:
        answer = input("Do you want to go to the next room? Type 'yes' or 'no': ").strip().lower()
        if answer == "yes":
            linebreak()
            play_room(next_room)
        else:
            linebreak()
            play_room(current_room)
    else:
        linebreak()
        play_room(current_room)
```
